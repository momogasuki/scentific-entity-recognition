```python
import urllib.request
```

```python
url = 'https://aclanthology.org/events/aacl-2020/'
urllib.request.urlopen(url).read().decode('utf-8')
```

```python
urllib.request.urlretrieve("https://aclanthology.org/2020.aacl-main.pdf", "raw_pdf/2020.aacl-main.pdf")
```

```python
from PyPDF2 import PdfReader
```

```python
reader = PdfReader("raw_pdf/2020.aacl-main.pdf")
# page = reader.pages[0]
# print(page.extract_text())
text = ''
for page in reader.pages:
    text += '\n' + page.extract_text()
```

```python
reader.pages[40].extract_text().split('\n')
```

```python
state = 0
processed_text = ''
for line in text.split('\n'):
    if line.lower() == 'references': state = 0
    elif (line[:2] == 'A ' and len(line) < 20) or line.lower() == 'abstract': state = 1
    if state == 0: continue
    if line[-1] == '-':
        processed_text += line[:-1]
    else:
        processed_text += line + ' '
```

```python
from spacy.lang.en import English
nlp = English()
tokenizer = nlp.tokenizer
```

```python
# list(tokenizer(processed_text))
with open('data.txt', 'w') as f:
    for token in tokenizer(processed_text.encode('utf-8', 'replace').decode()):
        f.write('{} '.format(token))
        if str(token) == '.': f.write('\n')
```
