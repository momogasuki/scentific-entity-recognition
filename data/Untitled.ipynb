```python
import urllib.request
```

```python
urllib.request.urlretrieve("https://aclanthology.org/2020.aacl-main.1.pdf", "2020.aacl-main.1.pdf")
```

```python
from PyPDF2 import PdfReader
```

```python
reader = PdfReader("2020.aacl-main.1.pdf")
page = reader.pages[0]
print(page.extract_text())
```

```python
from spacy.lang.en import English
nlp = English()
tokenizer = nlp.tokenizer
```

```python
list(tokenizer(page.extract_text()))
```
